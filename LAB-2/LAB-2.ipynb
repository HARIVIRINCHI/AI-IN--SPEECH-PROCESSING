A1. Find the first derivative of your speech signal with finite difference method. Listen to the first 
derivative signal and the original speech signal.

```python
import librosa
import librosa.display
import matplotlib.pyplot as plot
```

```python
filename = 'Hari1.wav'
```

```python
import IPython.display as ipdisp
```

```python
ipdisp.Audio(filename)
```

```python
y, sr = librosa.load(filename)
librosa.display.waveshow(y)
plot.title('Speech Signal Waveform')
plot.xlabel('Time')
plot.ylabel('Amplitude') 
plot.show()
```

```python
import numpy as np
from IPython.display import Audio
first_derivative = np.diff(y)
Audio(first_derivative, rate=sr)
```

```python
time_derivative = np.linspace(0, len(first_derivative)/sr, len(first_derivative))
plot.figure(figsize=(10, 4))
plot.plot(time_derivative, first_derivative)
plot.title('First Derivative of Speech Signal')
plot.xlabel('Time (s)')
plot.ylabel('Amplitude')
plot.show()
```

A2. Detect the points of zero crossing in the first derivative signal. Compare the average length 
between two consecutive zero crossings for speech and silence regions. Observe the pattern.

```python
zero_crossings_indices = np.where(np.diff(np.sign(first_derivative)))[0]
print("zero_crossings_indices:",zero_crossings_indices)
zero_crossings_diff = np.diff(zero_crossings_indices)
print("zero_crossings_diff:",zero_crossings_diff)
average_length = np.mean(zero_crossings_diff) / sr
print("Average length between two consecutive zero crossings (seconds):", average_length)
```

```python
plot.figure(figsize=(10, 4))
plot.plot(time_derivative, first_derivative, label='First Derivative')
plot.scatter(time_derivative[zero_crossings_indices], first_derivative[zero_crossings_indices], color='cyan', label='Zero Crossings')
plot.title('First Derivative of Speech Signal with Zero Crossings')
plot.xlabel('Time (s)')
plot.ylabel('Amplitude')
plot.legend()
plot.show()
```

A3. Speak 5 of your favorite words. Observe the length of the speech signals. Compare the lengths of 
your spoken words with those of your project team-mate.

```python
mine = 'lab2-A3-mine.wav'
teammate = 'lab2-A3-teammates.wav'
import os
mine_y, mine_sr = librosa.load(mine)
teammate_y, teammate_sr = librosa.load(teammate)
mine_duration = librosa.get_duration(y=mine_y, sr=mine_sr)
teammate_duration = librosa.get_duration(y=teammate_y, sr=teammate_sr)
print("Your spoken words duration:", mine_duration, "seconds")
print("Teammate's spoken words duration:", teammate_duration, "seconds")
```
